## how to add 1 to every number in table

```python
def add1_tab (tab):
"""Adds 1 to every number in tab,
a 2D list of numbers."""
for outer_idx in range(len(tab)):
for inner_idx in range(len(tab[outer_idx])):
old_val = tab[outer_idx][inner_idx]
tab[outer_idx][inner_idx] = old_val + 1
```

# TOM TAT

▪ For loops & ranges: Iterate items with for x in seq; use
range(start, step) for counted iterations.
▪ Accumulator pattern: Initialize once, e.g., total=0,
out=[], update inside the loop to build results.
▪ Items vs. indices: Prefer iterating items; use indices only for
in-place mutation or when positions matter
▪ Nested lists & loops: Model tables as lists-of-lists; traverse
for row in table: for cell in row.
▪ Tabular data: Aggregate, filter, and map without unintended
mutation; handle ragged rows and aliasing carefully.

```python
def matranchuyenvi(tab):
    for hang in range (len(tab)):
        for cot in range (len(tab[hang])):
            tab[hang][cot]=tab[cot][hang]
    print(tab)
a=[[1,2],[2,3],[3,4]]
matranchuyenvi(a)
```
