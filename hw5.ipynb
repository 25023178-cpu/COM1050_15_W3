```python
n=int(input())
def snt(n):
    if n < 2:
        return "khong phai snt"
    for i in range(2, int(n**0.5) + 1): 
        if n % i == 0:
            return "khong phai snt"
            
    return "la snt"
snt(n)
```

```python
a,b=map(int,input().split())
toantu=input()
def calc(a,b,toantu):
    if toantu=="+":
        print(a+b)
    elif toantu=="-":
        print(a-b)
    elif toantu=="*":
        print(a*b)
    elif toantu=="/":
        print(a/b)
    else:
        return "may ngu a"
calc(a,b,toantu)
```

# A10

```python
def anhxa():
    a,b=map(str,input().split())
    chu_a=list(a)
    chu_b=list(b)
    for i in range(len(a)+1):
        if chu_a[i]!=chu_b[i]:
            chu_a[i]=chu_b[i]
            chu_b[i]=chu_a[i]
        else:
            return 0
        return 1
anhxa()
            
    
```
