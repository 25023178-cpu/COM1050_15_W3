# A6

```python
def tongnguyento():
    a,b = map(int, input().split())
    tong_snt = 0
    
    for i in range(a, b + 1):
        if i < 2:
            return 
            
        for n in range(2, int(i**0.5) + 1): 
            if i % n == 0:
                break
        else:
            tong_snt += i
            
    print(tong_snt)

tongnguyento()
```

# A7

```python
def uocnguyento():
    n = int(input())
    
    ntt = 1
    d = 2
    
    while d ** 2 <= n:
        if n % d == 0:
            ntt = d
            while n % d == 0:
                n //= d
        d += 1
        
    if n > 1:
        ntt = n
        
    print(ntt)

uocnguyento()
```

# A9

```python
n = int(input())

chinhphuong = [
    str(k*k)
    for k in range(1, int(n**0.5) + 1)
    if len(set(str(k*k))) == len(str(k*k))
]

print('  '.join(chinhphuong) if chinhphuong else '(no number)')
```
