# BAI 13

```python
def hocluc():
    diem = float(input("Nhập điểm trung bình của học sinh: "))
    if diem<0 or diem>10:
       return hocluc()
    if diem >= 8.0:
        print("Học lực: Giỏi")
    elif diem >= 6.5:
        print("Học lực: Khá")
    elif diem >= 5.0:
        print("Học lực: Trung bình")
    else:
        print("Học lực: Yếu")
hocluc()
```
